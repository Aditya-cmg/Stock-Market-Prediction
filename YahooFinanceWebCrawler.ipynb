```python
import time
import datetime
import pandas as pd

#from - 14/12/2015 23:59 - 11/2/2021 23:59

ticker = '%5EGSPC'
period1 = int(time.mktime(datetime.datetime(2015, 12, 14, 23, 59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2021, 11, 2, 23, 59).timetuple()))
interval = '1d' # 1d, 1m

query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

df = pd.read_csv(query_string)
# print(df)
df.to_csv('C:\\Semester materials\\Spring Sem 2021 3rd 1 sem\\IBM 312\\GSPC.csv')
```
